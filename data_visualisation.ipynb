```python
# Visualising Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
n_samples, n_features = iris.data.shape
print("Number of features : {}".format(n_features))
print("Number of samples : {}".format(n_samples))
```

```python
print(iris.data.shape)
print(iris.target.shape)
```

```python
print(iris.target) #denotes the target labels
```

```python
import numpy as np

np.bincount(iris.target) #number of training data
```

```python
print(iris.target_names)
# 0 setosa
# 1 versicolor
# 2 virginica
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x_index = 2
y_index = 0
colors = ['blue','green','red']

for label,color in zip(range(len(iris.target_names)), colors):
    plt.scatter(iris.data[iris.target == label, x_index],
                iris.data[iris.target == label, y_index],
                label = iris.target_names[label],
                c = color)
    
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.legend(loc='upper left')
plt.show()
```

```python
# Visualising digits Datasets
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.keys()
```

```python
n_sample, n_features = digits.data.shape
print("number of features {}".format(n_features))
print("number of samples {}".format(n_samples))
```

```python
print(digits.data.shape)
print(digits.target.shape)
```

```python
print(digits.data[0]) #first data member
print(digits.target)  #classes
```

```python
print(digits.target_names)
print(digits.images.shape)
```

```python
# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```
